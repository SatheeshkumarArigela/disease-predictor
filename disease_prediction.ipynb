```python
import pandas as pd
dataset=pd.read_csv('https://raw.githubusercontent.com/maddydevgits/ML-Dev-Hackathon-2022-GPCET-Kurnool/main/day3/disease-prediction/Training.csv')
```

```python
dataset
```

```python
dataset.columns
```

```python
X=dataset.iloc[:,:-1].values
```

```python
X
```

```python
Y=dataset.iloc[:,-1].values
```

```python
Y
```

```python
Y=pd.get_dummies(Y)
```

```python
Y
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X,Y)
```
